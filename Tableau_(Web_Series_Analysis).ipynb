```python
# Task 6: Platform-Wise Average IMDB Rating by Genre
#Goal: Analyze which genres perform best on which streaming platforms in terms of average IMDB ratings.
#Visualization: Heatmap (Rows = Genre, Columns = Platform, Color = Avg. IMDB Rating)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "/content/Web_Series_Dataset.xlsx"   # update path if needed
data = pd.read_excel(file_path, sheet_name="Sheet1")

# Check column names
print(data.columns)

# Assuming columns: 'Genre', 'Streaming Platform', 'IMDB Rating'
# Replace with exact names from your dataset
pivot_table = data.pivot_table(
    index="Genre",
    columns="Platform",
    values="IMDB Rating",
    aggfunc="mean"
)

# Plot heatmap
plt.figure(figsize=(12,8))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="YlGnBu", linewidths=0.5)

plt.title("Platform-Wise Average IMDB Rating by Genre", fontsize=14)
plt.xlabel("Streaming Platform")
plt.ylabel("Genre")
plt.show()
```

```python
# Task 7: Episodes Per Season Analysis
#Goal: Compute and compare the average number of episodes per season for each web series.
#Visualization: Bar Chart or Bubble Chart (Series on X-axis, Episodes/Season ratio as value)
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "/content/Web_Series_Dataset.xlsx"   # update path if needed
data = pd.read_excel(file_path, sheet_name="Sheet1")

# Check column names
print(data.columns)

# Assuming columns: 'Title', 'Episodes', 'Seasons'
# Replace with exact column names from your dataset
data["Episodes per Season"] = data["Episodes"] / data["Seasons"]

# Sort by highest episodes/season
eps_per_season = data.sort_values(by="Episodes per Season", ascending=False)

# --- Option 1: Bar Chart ---
plt.figure(figsize=(12,6))
plt.bar(eps_per_season["Title"], eps_per_season["Episodes per Season"], color="skyblue")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Average Episodes per Season")
plt.title("Episodes Analysis (Bar Chart)")
plt.show()

# --- Option 2: Bubble Chart ---
plt.figure(figsize=(12,6))
plt.scatter(
    eps_per_season["Title"],
    eps_per_season["Episodes per Season"],
    s=eps_per_season["Episodes per Season"]*20,  # Bubble size scaled
    alpha=0.6,
    color="orange"
)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Average Episodes per Season")
plt.title("Episodes Per Season Analysis (Bubble Chart)")
plt.show()
```

```python
#Task 9: Identify the Most Consistent Genre in Ratings
#Goal: Find the genre with the lowest variance in IMDB ratings (i.e., most consistent viewer reception).
#Visualization: Table or Bar Chart with genre and rating variance
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "/content/Web_Series_Dataset.xlsx"   # update path if needed
data = pd.read_excel(file_path, sheet_name="Sheet1")

# Check column names
print(data.columns)

# Assuming columns: 'Genre' and 'IMDB Rating'
# Replace with exact names from your dataset
genre_variance = data.groupby("Genre")["IMDB Rating"].var().sort_values()

# Display as table
print("Genre Rating Variance:")
print(genre_variance)

# Plot bar chart
plt.figure(figsize=(10,6))
genre_variance.plot(kind="bar", color="skyblue")
plt.ylabel("Variance in IMDB Ratings")
plt.title("Variance of IMDB Ratings by Genre (Lower = More Consistent)")
plt.xticks(rotation=45, ha="right")
plt.show()

# Most consistent genre
most_consistent = genre_variance.idxmin()
print(f"\n✅ The most consistent genre (lowest variance) is: {most_consistent}")
```

```python
#Task 10: Highlight Web Series with Above-Average Ratings in Each Genre
#Goal: For each genre, list web series that have an IMDB rating higher than the genre’s average.
#Visualization: Filtered Bar Chart or Table view, with conditional formatting on ratings
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "/content/Web_Series_Dataset.xlsx"   # update path if needed
data = pd.read_excel(file_path, sheet_name="Sheet1")

# Check column names
print(data.columns)

# Assuming columns: 'Title', 'Genre', 'IMDB Rating'
# Replace with actual names from your dataset

# Compute average rating per genre
genre_avg = data.groupby("Genre")["IMDB Rating"].mean().reset_index()
genre_avg.rename(columns={"IMDB Rating": "Genre_Avg_Rating"}, inplace=True)

# Merge back with original dataset
merged = pd.merge(data, genre_avg, on="Genre", how="left")

# Filter web series with rating above genre average
above_avg = merged[merged["IMDB Rating"] > merged["Genre_Avg_Rating"]]

# --- Table Output ---
print("✅ Web Series with Above-Average Ratings in Each Genre:")
print(above_avg[["Title", "Genre", "IMDB Rating", "Genre_Avg_Rating"]].sort_values(by=["Genre", "IMDB Rating"], ascending=[True, False]))

# --- Bar Chart Visualization ---
plt.figure(figsize=(12,6))
sns.barplot(
    data=above_avg,
    x="Genre",
    y="IMDB Rating",
    hue="Title",
    dodge=False
)
plt.axhline(y=data["IMDB Rating"].mean(), color="red", linestyle="--", label="Overall Avg Rating")
plt.title("Web Series with Above-Average Ratings in Each Genre")
plt.ylabel("IMDB Rating")
plt.xticks(rotation=45, ha="right")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
