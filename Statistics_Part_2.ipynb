1. What is hypothesis testing in statistics?
 - Hypothesis testing in statistics is a method used to make decisions or inferences about population parameters based on sample data. It helps you determine whether there is enough evidence in a sample to support a specific claim about a population.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
 - Null Hypothesis (H₀) is the default or starting assumption. It states that there is no effect, no difference, or no relationship. We test against this hypothesis to see if it can be rejected.
  - Alternative Hypothesis (H₁ or Ha) is what we want to find evidence for. It states that there is an effect, a difference, or a relationship. If the data contradicts the null hypothesis strongly enough, you accept the alternative.

3. What is the significance level in hypothesis testing, and why is it important?
 - The significance level in hypothesis testing, often denoted by α (alpha), is a threshold used to determine whether the null hypothesis (H₀) should be rejected.

 ❗ Why It’s Important
  1. Controls the Risk of False Positives (Type I Error)

  - A Type I error is when you say there’s an effect when there isn’t one.

  - Example: Claiming a new drug works better when it actually doesn’t.

   2. Determines the Critical Region for Your Test

   - If your p-value is less than or equal to α, you reject the null hypothesis.

  - If your p-value is greater than α, you fail to reject the null hypothesis.

  3. Sets the Standard of Evidence

  - A smaller α (like 0.01) means you're demanding stronger evidence before rejecting H₀.

  - A larger α (like 0.10) means you’re willing to take more risk of being wrong.

4. What does a P-value represent in hypothesis testing?
 - The p-value is the probability of observing your sample results, or something more extreme, assuming the null hypothesis (H₀) is true.
Small p-value (e.g., < 0.05)	Strong evidence against H₀ → You reject the null hypothesis.
Large p-value (e.g., > 0.05)	Weak evidence against H₀ → You fail to reject the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
 - The p-value is the probability of getting results as extreme as (or more extreme than) your observed data, assuming the null hypothesis (H₀) is true.

 ✅ Steps to Interpret a P-value
Compare the p-value to the significance level (α):

Common α values: 0.05, 0.01, 0.10

Make a decision:

If p ≤ α → Reject the null hypothesis (results are statistically significant)

If p > α → Fail to reject the null hypothesis (results are not statistically significant)

6. What are Type 1 and Type 2 errors in hypothesis testing?
 - Type I and Type II errors are two kinds of mistakes that can occur in hypothesis testing when making decisions based on sample data. Type I errors means false positive and Type II errors means false negative.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
 -  One-Tailed Test
Tests for an effect in one specific direction only.

You use it when you have a directional hypothesis (e.g., you believe a mean is greater than or less than a certain value).

Example:
H₀ (null hypothesis): μ ≤ 100

H₁ (alternative hypothesis): μ > 100

You're only interested in whether the value is greater than 100, not less.

Characteristics:
Entire alpha level (e.g., 0.05) is in one tail of the distribution.

More powerful in detecting an effect in the specified direction.

Not appropriate if the effect could reasonably go in either direction.

🔹 Two-Tailed Test
Tests for an effect in either direction.

You use it when you have a non-directional hypothesis (e.g., you believe a mean is different from a certain value, but not sure if it's higher or lower).

Example:
H₀: μ = 100

H₁: μ ≠ 100

You're testing if the mean is either greater than or less than 100.

Characteristics:
Alpha level is split between both tails (e.g., 0.025 in each tail if α = 0.05).

More conservative, less likely to find significance unless the effect is strong.

Appropriate when no prior reason exists to expect a change in a specific direction.

8. What is the Z-test, and when is it used in hypothesis testing?
 - The Z-test is a type of statistical test used in hypothesis testing to determine whether there is a significant difference between sample and population means (or between two sample means), assuming the population follows a normal distribution and the population standard deviation (σ) is known.

🧪 When is the Z-Test Used?
You use a Z-test in the following situations:

 1.The sample size is large (typically n ≥ 30).

 2.The population standard deviation (σ) is known.

 3.The data is approximately normally distributed (especially for small samples).

We're testing:

A sample mean against a known population mean.

The difference between two means (in some cases).

A sample proportion against a population proportion.

  When to Use the T-Distribution Instead of the Normal Distribution-
  1.Population standard deviation is unknown
  2.Sample size is small (n < 30)
  3.Population distribution is approximately normal or unknown.

11. What is the difference between a Z-test and a T-test?
 - The Z-test and T-test are both statistical methods used to determine whether there is a significant difference between sample statistics and population parameters, but they differ primarily in their assumptions and appropriate use cases. The Z-test is used when the population standard deviation (σ) is known and the sample size is large (typically n ≥ 30). It relies on the standard normal distribution, which has a fixed shape. In contrast, the T-test is used when the population standard deviation is unknown and especially when the sample size is small (typically n < 30). It uses the Student’s t-distribution, which has heavier tails than the normal distribution to account for the extra uncertainty from estimating σ using the sample standard deviation. As the sample size increases, the t-distribution approaches the normal distribution, making the T-test increasingly similar to the Z-test. Therefore, the key difference lies in whether the population standard deviation is known and the size of the sample.

13. What is the relationship between Z-test and T-test in hypothesis testing?
 - The Z-test and T-test are closely related statistical tools used in hypothesis testing to assess whether a sample mean (or difference between means) significantly differs from a known or hypothesized value. Both tests follow the same core logic — they compare a test statistic to a critical value derived from a theoretical distribution — but they differ based on assumptions about the data.

14. What is a confidence interval, and how is it used to interpret statistical results?
 - A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (such as the mean or proportion) with a certain level of confidence.

It reflects the uncertainty or variability in your estimate due to sampling.

🔍 How Confidence Intervals Are Used to Interpret Statistical Results
A confidence interval helps you understand the precision and reliability of your estimate:

A narrow interval suggests a more precise estimate.

A wide interval indicates more uncertainty.

The confidence level (e.g., 95%) means that if the same study were repeated many times, about 95% of the resulting confidence intervals would contain the true population parameter.

👉 Interpretation Example:
If you calculate a 95% confidence interval for the average height of a population as (165 cm, 175 cm), you can interpret it as:

“We are 95% confident that the true average height of the population lies between 165 cm and 175 cm.”

16. How is Bayes' Theorem used in statistics, and what is its significance?
 - Bayes' Theorem is a fundamental rule in probability and statistics that describes how to update the probability of a hypothesis based on new evidence or data. It allows statisticians to combine prior knowledge with observed data to calculate a revised (posterior) probability.

⚡ Significance of Bayes' Theorem
  1.Incorporates prior knowledge: Unlike classical (frequentist) statistics, Bayesian methods formally include prior beliefs or information.

  2.Dynamic updating: Probabilities are updated as new evidence accumulates, making it adaptive.

  3.Provides full probability distributions: Rather than just point estimates, Bayesian methods yield a complete distribution reflecting uncertainty.

  4.Widely applicable: Used in fields ranging from genetics and medicine to finance and AI.

17.  What is the Chi-square distribution, and when is it used?
 - The Chi-square (χ²) distribution is a continuous probability distribution that describes the distribution of a sum of the squares of independent standard normal random variables. It is positively skewed and defined only for non-negative values (χ² ≥ 0).
 📊 When is the Chi-square Distribution Used?
  1.The Chi-square distribution is primarily used in hypothesis testing and confidence interval estimation involving categorical data or variance estimates:

  2.Chi-square Test for Independence:
Tests whether two categorical variables are independent (e.g., testing if smoking status is independent of lung disease occurrence in a population).

  3.Chi-square Goodness-of-Fit Test:
Tests if an observed frequency distribution fits an expected distribution (e.g., checking if a dice is fair by comparing observed roll frequencies to expected frequencies).

  4.Test for Population Variance:
Used to test hypotheses about the variance or standard deviation of a normally distributed population when the mean is known.

🔍 When is the F-Distribution Used?
The F-distribution is primarily used in hypothesis testing situations involving the comparison of variances or multiple group means, such as:

Analysis of Variance (ANOVA):

To test if there are significant differences between the means of three or more groups.

ANOVA compares the variance between groups to the variance within groups using an F-statistic.

20. What is an ANOVA test, and what are its assumptions?
 - ANOVA stands for Analysis of Variance. It is a statistical method used to compare the means of three or more groups to see if at least one group mean is significantly different from the others. Rather than comparing means pairwise, ANOVA tests all groups simultaneously, helping to control the overall error rate.
 ✅ Assumptions of ANOVA
 1.Independence of observations:
Each group’s samples are independent of the others.

 2.Normality:
The data in each group should be approximately normally distributed.

 3.Homogeneity of variances (Homoscedasticity):
The variances in each group should be roughly equal.

 -  Differents types of ANOVA tests are

 4. Multivariate ANOVA (MANOVA)

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define discrete uniform distribution parameters
low = 1     # inclusive
high = 7    # exclusive (acts like a 6-sided die: 1 to 6)

# Create the distribution
dist = randint(low, high)

# Possible outcomes
x = np.arange(low, high)

# PMF values
pmf_vals = dist.pmf(x)

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(x, pmf_vals, color='lightgreen', edgecolor='black')
plt.title('Discrete Uniform Distribution PMF')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
        x (int): The outcome (0 or 1).
        p (float): The probability of success (x = 1), 0 ≤ p ≤ 1.

    Returns:
        float: The probability of the given outcome.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p if x == 1 else 1 - p

# Example usage
print("P(X=0):", bernoulli_pmf(0, 0.3))
print("P(X=1):", bernoulli_pmf(1, 0.3))

```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters of the discrete uniform distribution
low = 1   # inclusive
high = 7  # exclusive, so values are from 1 to 6

# Create the discrete uniform distribution
dist = randint(low, high)

# Possible values (support)
x = np.arange(low, high)

# Compute the CDF values
cdf_vals = dist.cdf(x)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_vals, where='post', color='blue', label='CDF')
plt.scatter(x, cdf_vals, color='red', zorder=5)
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.ylim(0, 1.1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Avoid division by zero
    if std == 0:
        raise ValueError("Standard deviation is zero, cannot compute Z-scores.")

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='none', color='blue', alpha=0.6)
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.axhline(1, color='red', linestyle='--', linewidth=1, label='Z = ±1')
    plt.axhline(-1, color='red', linestyle='--', linewidth=1)
    plt.axhline(2, color='orange', linestyle='--', linewidth=1, label='Z = ±2')
    plt.axhline(-2, color='orange', linestyle='--', linewidth=1)
    plt.title('Z-Scores of Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-Score')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    return z_scores
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the non-normal (uniform) distribution
population = np.random.uniform(low=0, high=100, size=100000)

# Step 2: Define parameters for CLT simulation
sample_size = 30        # number of data points in each sample
num_samples = 1000      # how many sample means to generate

# Step 3: Generate sample means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Step 4: Plot histogram of sample means
plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=30, color='lightblue', edgecolor='black', density=True)
plt.title(f'CLT Demonstration: Sampling Distribution of Mean\n(sample size={sample_size}, num samples={num_samples})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, point):
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero, cannot compute Z-score.")

    z = (point - mean) / std
    return z

def plot_z_score_on_standard_normal(z):
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, label='Standard Normal PDF', color='blue')
    plt.fill_between(x, 0, pdf, where=(x <= z), color='skyblue', alpha=0.5, label=f'P(Z ≤ {z:.2f})')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

    plt.title('Z-Score on Standard Normal Distribution')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Example dataset
    data = np.array([12, 15, 14, 16, 13, 15, 17, 14, 13, 15])
    point = 18  # Data point to calculate Z-score for

    z_score = calculate_z_score(data, point)
    print(f'Z-score for data point {point}: {z_score:.2f}')

    plot_z_score_on_standard_normal(z_score)
```

```python
#15.  Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate standard error
    std_error = pop_std / np.sqrt(n)

    # Calculate Z-statistic
    z_stat = (sample_mean - pop_mean) / std_error

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Decision
    reject_null = p_value < alpha

    return z_stat, p_value, reject_null

# Example usage
if __name__ == "__main__":
    # Sample data
    sample_data = np.array([102, 98, 100, 105, 97, 101, 99, 103, 100, 104])

    # Known population parameters
    population_mean = 100
    population_std = 5  # population standard deviation (known)

    # Significance level
    alpha = 0.05

    z_stat, p_val, reject = z_test(sample_data, population_mean, population_std, alpha)

    print(f"Sample mean: {np.mean(sample_data):.2f}")
    print(f"Z-statistic: {z_stat:.3f}")
    print(f"P-value: {p_val:.4f}")
    print(f"Reject null hypothesis? {'Yes' if reject else 'No'}")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean

    # t-critical value for two-tailed test
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of error
    moe = t_crit * std_err

    # Confidence interval
    ci_lower = mean - moe
    ci_upper = mean + moe

    return mean, (ci_lower, ci_upper)

# Example usage
if __name__ == "__main__":
    sample_data = np.array([12, 15, 14, 16, 13, 15, 17, 14, 13, 15])

    confidence_level = 0.95
    mean, (lower, upper) = confidence_interval(sample_data, confidence=confidence_level)

    print(f"Sample mean: {mean:.2f}")
    print(f"{int(confidence_level*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {int(confidence_level*100)}% confident that the population mean lies between {lower:.2f} and {upper:.2f}.")
```

```python
#PRACTICAL PART 2
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the result.
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Performs a two-tailed Z-test for a sample mean against a population mean.

    Parameters:
    - sample: array-like sample data
    - pop_mean: known population mean
    - pop_std: known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: calculated Z-statistic
    - p_value: two-tailed p-value
    - reject_null: boolean indicating if null hypothesis is rejected
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    std_error = pop_std / np.sqrt(n)

    z_stat = (sample_mean - pop_mean) / std_error
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed p-value

    reject_null = p_value < alpha

    return z_stat, p_value, reject_null

# Example usage
if __name__ == "__main__":
    # Sample data
    sample_data = np.array([102, 98, 100, 105, 97, 101, 99, 103, 100, 104])

    # Known population parameters
    population_mean = 100
    population_std = 5  # population std deviation (assumed known)
    alpha = 0.05  # significance level

    z_stat, p_val, reject = z_test(sample_data, population_mean, population_std, alpha)

    print(f"Sample mean: {np.mean(sample_data):.2f}")
    print(f"Z-statistic: {z_stat:.3f}")
    print(f"P-value: {p_val:.4f}")
    print(f"Reject null hypothesis? {'Yes' if reject else 'No'}")

    if reject:
        print(f"Conclusion: At the {alpha*100}% significance level, there is enough evidence to reject the null hypothesis that the sample mean equals the population mean.")
    else:
        print(f"Conclusion: At the {alpha*100}% significance level, there is not enough evidence to reject the null hypothesis.")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    std_err = pop_std / np.sqrt(n)

    z_stat = (sample_mean - pop_mean) / std_err
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    reject_null = p_value < alpha

    return z_stat, p_value, reject_null

def plot_z_test(z_stat, alpha=0.05):
    # Critical values for two-tailed test
    z_crit = norm.ppf(1 - alpha/2)

    # Create x values for standard normal
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_crit), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, 0, y, where=(x >= z_crit), color='red', alpha=0.3)

    # Mark the observed z_stat
    plt.axvline(z_stat, color='blue', linestyle='--', label=f'Observed Z = {z_stat:.3f}')

    plt.title('Two-tailed Z-Test Decision Regions')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Example usage
if __name__ == "__main__":
    np.random.seed(42)
    sample = np.random.normal(loc=102, scale=15, size=50)
    population_mean = 100
    population_std = 15
    alpha = 0.05

    z_stat, p_val, reject = two_tailed_z_test(sample, population_mean, population_std, alpha)

    print(f"Sample mean: {np.mean(sample):.2f}")
    print(f"Z-statistic: {z_stat:.3f}")
    print(f"P-value: {p_val:.4f}")
    print("Reject null hypothesis." if reject else "Fail to reject null hypothesis.")

    plot_z_test(z_stat, alpha)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(pop_mean, pop_std, sample_size, alpha, alt_mean):
    """
    Visualize Type I and Type II errors for a one-sample Z-test.

    Parameters:
    - pop_mean: mean under the null hypothesis H0
    - pop_std: population standard deviation
    - sample_size: size of the sample
    - alpha: significance level (Type I error rate)
    - alt_mean: true mean under alternative hypothesis H1
    """
    # Standard error
    se = pop_std / np.sqrt(sample_size)

    # Critical z-value for two-tailed test
    z_crit = norm.ppf(1 - alpha/2)

    # Critical values in terms of sample mean
    crit_low = pop_mean - z_crit * se
    crit_high = pop_mean + z_crit * se

    # Define range for plotting
    x = np.linspace(pop_mean - 4*se, pop_mean + 4*se, 1000)

    # Null distribution (H0)
    null_pdf = norm.pdf(x, loc=pop_mean, scale=se)
    # Alternative distribution (H1)
    alt_pdf = norm.pdf(x, loc=alt_mean, scale=se)

    plt.figure(figsize=(12, 6))

    # Plot null hypothesis distribution
    plt.plot(x, null_pdf, label='Null Hypothesis (H0)', color='blue')
    # Plot alternative hypothesis distribution
    plt.plot(x, alt_pdf, label='Alternative Hypothesis (H1)', color='green')

    # Shade Type I error regions (tails under H0 beyond critical values)
    x_fill_left = np.linspace(pop_mean - 4*se, crit_low, 200)
    x_fill_right = np.linspace(crit_high, pop_mean + 4*se, 200)
    plt.fill_between(x_fill_left, 0, norm.pdf(x_fill_left, pop_mean, se), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x_fill_right, 0, norm.pdf(x_fill_right, pop_mean, se), color='red', alpha=0.3)

    # Shade Type II error region (area under H1 curve within acceptance region)
    x_accept = np.linspace(crit_low, crit_high, 400)
    plt.fill_between(x_accept, 0, norm.pdf(x_accept, alt_mean, se), color='orange', alpha=0.3, label='Type II Error (β)')

    # Plot critical value lines
    plt.axvline(crit_low, color='black', linestyle='--', label='Critical Values')
    plt.axvline(crit_high, color='black', linestyle='--')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    # Calculate Type II error (β)
    beta_left = norm.cdf(crit_low, loc=alt_mean, scale=se)
    beta_right = 1 - norm.cdf(crit_high, loc=alt_mean, scale=se)
    beta = beta_left + beta_right

    print(f"Type I error rate (α): {alpha}")
    print(f"Type II error rate (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage:
if __name__ == "__main__":
    visualize_type1_type2_errors(
        pop_mean=100,
        pop_std=15,
        sample_size=30,
        alpha=0.05,
        alt_mean=110
    )
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent two-sample t-test.

    Parameters:
    - sample1: array-like, first sample data
    - sample2: array-like, second sample data
    - alpha: significance level (default 0.05)

    Returns:
    - t_stat: calculated t-statistic
    - p_value: two-tailed p-value
    - reject_null: True if null hypothesis rejected, else False
    """
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)  # Welch's t-test (does not assume equal variances)
    reject_null = p_value < alpha

    return t_stat, p_value, reject_null

# Example usage
if __name__ == "__main__":
    np.random.seed(0)
    # Generate two independent samples
    group1 = np.random.normal(loc=50, scale=5, size=30)
    group2 = np.random.normal(loc=53, scale=5, size=30)

    alpha = 0.05
    t_stat, p_val, reject = independent_t_test(group1, group2, alpha)

    print(f"Group 1 mean: {np.mean(group1):.2f}")
    print(f"Group 2 mean: {np.mean(group2):.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_val:.4f}")
    print("Reject null hypothesis." if reject else "Fail to reject null hypothesis.")

    if reject:
        print(f"Conclusion: There is a significant difference between the two group means at the {alpha*100}% significance level.")
    else:
        print(f"Conclusion: There is no significant difference between the two group means at the {alpha*100}% significance level.")
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test(sample_before, sample_after, alpha=0.05):
    """
    Perform a paired sample t-test.

    Parameters:
    - sample_before: array-like data before treatment
    - sample_after: array-like data after treatment
    - alpha: significance level

    Returns:
    - t_stat: t-statistic
    - p_value: two-tailed p-value
    - reject_null: True if null hypothesis rejected
    """
    t_stat, p_value = ttest_rel(sample_before, sample_after)
    reject_null = p_value < alpha
    return t_stat, p_value, reject_null

def plot_paired_data(sample_before, sample_after):
    n = len(sample_before)
    plt.figure(figsize=(8,6))

    # Plot points and lines connecting paired samples
    for i in range(n):
        plt.plot([1, 2], [sample_before[i], sample_after[i]], marker='o', color='gray', alpha=0.6)

    # Plot group means
    means = [np.mean(sample_before), np.mean(sample_after)]
    plt.plot([1, 2], means, marker='o', color='red', markersize=10, label='Group Means')

    plt.xticks([1, 2], ['Before', 'After'])
    plt.ylabel('Value')
    plt.title('Paired Sample Data Comparison')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Example usage
if __name__ == "__main__":
    np.random.seed(42)
    # Simulate paired samples (e.g., before and after treatment)
    sample_before = np.random.normal(loc=75, scale=10, size=20)
    sample_after = sample_before + np.random.normal(loc=5, scale=8, size=20)  # some improvement

    t_stat, p_val, reject = paired_t_test(sample_before, sample_after)

    print(f"Before mean: {np.mean(sample_before):.2f}")
    print(f"After mean: {np.mean(sample_after):.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_val:.4f}")
    print("Reject null hypothesis." if reject else "Fail to reject null hypothesis.")

    plot_paired_data(sample_before, sample_after)
```

```python
#8. Simulate data and perform both Z-test and T-test, then compare the results using Python.
import numpy as np
from scipy.stats import norm, t, ttest_1samp

def one_sample_z_test(sample, pop_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    std_err = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / std_err
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    return z_stat, p_value

def one_sample_t_test(sample, pop_mean):
    t_stat, p_value = ttest_1samp(sample, pop_mean)
    return t_stat, p_value

if __name__ == "__main__":
    np.random.seed(123)

    # Known population parameters
    population_mean = 100
    population_std = 15  # known for Z-test

    # Simulate sample (mean slightly shifted)
    sample_size = 40
    sample = np.random.normal(loc=104, scale=population_std, size=sample_size)

    # Z-test (population std dev known)
    z_stat, z_p = one_sample_z_test(sample, population_mean, population_std)

    # T-test (population std dev unknown)
    t_stat, t_p = one_sample_t_test(sample, population_mean)

    print("=== One-sample Z-test ===")
    print(f"Sample mean: {np.mean(sample):.2f}")
    print(f"Z-statistic: {z_stat:.3f}")
    print(f"P-value: {z_p:.4f}")

    print("\n=== One-sample T-test ===")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {t_p:.4f}")

    alpha = 0.05
    print("\nInterpretation:")
    if z_p < alpha:
        print("Z-test: Reject null hypothesis (significant difference).")
    else:
        print("Z-test: Fail to reject null hypothesis.")

    if t_p < alpha:
        print("T-test: Reject null hypothesis (significant difference).")
    else:
        print("T-test: Fail to reject null hypothesis.")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - sample: array-like sample data
    - confidence: confidence level (default 0.95)

    Returns:
    - mean: sample mean
    - lower_bound: lower bound of the confidence interval
    - upper_bound: upper bound of the confidence interval
    """
    n = len(sample)
    mean = np.mean(sample)
    sem = np.std(sample, ddof=1) / np.sqrt(n)  # standard error of the mean

    # t-critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * sem
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return mean, lower_bound, upper_bound

# Example usage
if __name__ == "__main__":
    np.random.seed(1)
    sample_data = np.random.normal(loc=50, scale=8, size=40)
    conf_level = 0.95

    mean, lower, upper = confidence_interval(sample_data, conf_level)

    print(f"Sample mean: {mean:.2f}")
    print(f"{conf_level*100:.1f}% Confidence Interval: [{lower:.2f}, {upper:.2f}]")
    print(f"Interpretation: We are {conf_level*100:.1f}% confident that the true population mean lies within this interval.")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for the sample mean.

    Parameters:
    - sample: array-like sample data
    - confidence: confidence level (default 0.95)

    Returns:
    - margin_of_error: the margin of error for the confidence interval
    """
    n = len(sample)
    sem = np.std(sample, ddof=1) / np.sqrt(n)  # standard error of the mean
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)  # two-tailed t critical value
    moe = t_crit * sem
    return moe

# Example usage
if __name__ == "__main__":
    import random
    random.seed(42)
    sample_data = np.random.normal(loc=100, scale=15, size=50)
    conf_level = 0.95

    moe = margin_of_error(sample_data, conf_level)
    print(f"Margin of Error at {conf_level*100:.1f}% confidence level: ±{moe:.3f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior, likelihood, false_positive_rate):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(Disease), prior probability of the event
    - likelihood: P(Positive|Disease), probability of positive test given disease
    - false_positive_rate: P(Positive|NoDisease), probability of positive test given no disease

    Returns:
    - posterior: P(Disease|Positive), posterior probability
    """
    # Complement of prior
    prior_neg = 1 - prior

    # Total probability of testing positive
    marginal_prob = likelihood * prior + false_positive_rate * prior_neg

    # Posterior probability (Bayes' theorem)
    posterior = (likelihood * prior) / marginal_prob

    return posterior

if __name__ == "__main__":
    # Given values
    prior = 0.01                    # P(Disease)
    likelihood = 0.99               # P(Positive|Disease)
    false_positive_rate = 0.05      # P(Positive|NoDisease)

    posterior_prob = bayes_theorem(prior, likelihood, false_positive_rate)

    print(f"Posterior probability of having the disease given a positive test: {posterior_prob:.4f}")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
from scipy.stats import chi2_contingency

# Example data: Contingency table of two categorical variables
# For example, survey counts of preference by gender:
# Rows: Gender (Male, Female)
# Columns: Preference (Like, Dislike)
data = np.array([[30, 10],
                 [20, 40]])

chi2_stat, p_val, dof, expected = chi2_contingency(data)

print("Chi-square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_val)
print("Expected Frequencies:\n", expected)

alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis - variables are dependent (associated).")
else:
    print("Fail to reject the null hypothesis - variables are independent.")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for Chi-square test.

    Parameters:
    - observed: 2D numpy array of observed counts

    Returns:
    - expected: 2D numpy array of expected counts
    """
    # Sum over rows and columns
    row_sums = observed.sum(axis=1).reshape(-1, 1)
    col_sums = observed.sum(axis=0).reshape(1, -1)
    total = observed.sum()

    # Expected frequency formula: (row_sum * col_sum) / total
    expected = (row_sums @ col_sums) / total
    return expected

# Example usage
if __name__ == "__main__":
    observed_data = np.array([[30, 10],
                              [20, 40]])

    expected_freq = calculate_expected_frequencies(observed_data)
    print("Observed Frequencies:\n", observed_data)
    print("Expected Frequencies:\n", expected_freq)
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

# Observed counts for categories (example)
observed = np.array([50, 30, 20])

# Expected proportions or counts
# For example, expected uniform distribution over 3 categories
expected_proportions = np.array([1/3, 1/3, 1/3])

# Convert proportions to expected counts (scaled to total observed)
expected = expected_proportions * observed.sum()

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Observed distribution differs from expected.")
else:
    print("Fail to reject null hypothesis: Observed distribution fits expected.")
```
