1. What is a random variable in probability theory?
 - In probability theory, a random variable is a function that assigns a numerical value to each possible outcome in a sample space of a random experiment.

2. What are the types of random variables?
 - There are two main types of random variables in probability theory:

 1. Discrete Random Variable
Takes on a countable set of distinct values.

 - Usually arises from counting.

 - Values can be finite or countably infinite (like 0, 1, 2, ...).

 Examples:

- Number of heads in 3 coin tosses (values: 0, 1, 2, 3)

- Number of students in a classroom

- Roll of a die (values: 1 through 6)

  2. Continuous Random Variable
Takes on values from a continuous range (usually an interval on the real number line).

 - Usually arises from measuring.

 - Infinite number of possible values.

 Examples:

- Height of a person (e.g., 170.2 cm)

- Time taken to complete a task (e.g., 2.53 seconds)

- Temperature at noon

9. What is the Poisson distribution and where is it applied?
 - The Poisson distribution is a discrete probability distribution that models the number of events occurring within a fixed interval of time or space, given that:

The events occur independently.

The events happen with a constant average rate λ, often referred to as the rate parameter.

It is often used to model the count of events that happen in a given period or region, especially when those events are rare relative to the observation period.

🔹 Use Cases of Poisson Distribution
The Poisson distribution is particularly useful in situations where:

Events are rare and occur independently.

The rate of occurrence is constant over time or space.

Common examples include:

Call centers: Number of calls received in an hour.

Traffic flow: Number of cars passing a specific point on a road in a given period.

Accidents: Number of accidents occurring at an intersection within a week.

Queuing systems: Number of customers arriving at a bank in an hour.

Biology/Medicine: Number of mutations in a given DNA sequence.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
 - The CLT states that:
When you take sufficiently large random samples from any population (with a finite mean and variance), the distribution of the sample means will approximate a normal distribution — regardless of the original population's distribution.

🔹 Why CLT Is Critical in Statistics
 1.Justifies Use of the Normal Distribution

Enables us to use normal-based methods (like z-tests, confidence intervals) even with non-normal populations.

 2.Foundation for Inferential Statistics

It allows us to make inferences about population parameters using sample data.

 3.Basis for Many Statistical Tests

Hypothesis testing, confidence intervals, and control charts often rely on the CLT.

 4.Works with Unknown Populations

As long as the sample size is large and data is independent, the CLT holds.


14. How does the Central Limit Theorem relate to the normal distribution?
 - The Central Limit Theorem (CLT) directly connects to the normal distribution by explaining why and when the normal distribution can be used to approximate the distribution of sample means — even if the original population is not normal.

15. What is the application of Z statistics in hypothesis testing?
 - Z-statistics (or Z-scores) are widely used in hypothesis testing to determine how far a sample statistic (like a sample mean or proportion) deviates from a hypothesized population parameter, assuming the null hypothesis is true.
 When to Use Z-Tests
  - Testing a population mean when σ is known and n is large

  - Testing a population proportion

  - Comparing two proportions or two means (with known variances)

  - Applications in quality control, A/B testing, medical trials, etc.

19. What is the relationship between a Z-score and a confidence interval?
 - The Z-score and the confidence interval (CI) are closely related — the Z-score is used to calculate the margin of error in a confidence interval when the population standard deviation is known (or the sample size is large).

20. How are Z-scores used to compare different distributions?
 - 🔹 How Z-scores Are Used to Compare Different Distributions
Standardizing Data
Z-scores standardize data, transforming it into a common scale (standard normal distribution with a mean of 0 and a standard deviation of 1). This allows you to compare values that come from different distributions, even if those distributions have different units or different means and standard deviations.

Adjusting for Different Units and Scales
When comparing data from different sources (e.g., comparing test scores in two different subjects or sales figures from different stores), Z-scores help to eliminate the effects of different units of measurement or different variability. This way, you're comparing the relative position within each distribution rather than absolute values.

21. What are the assumptions for applying the Central Limit Theorem?
 - 🔹 Key Assumptions for the Central Limit Theorem (CLT)
Random Sampling

The data must be randomly sampled from the population. Each sample should be independent, meaning the selection of one data point should not affect the selection of another.

If the samples are not random or independent, the CLT may not apply.

Sample Size (n)

For the CLT to hold, the sample size must be large enough. A general rule of thumb is that a sample size of

If you're sampling without replacement, the sample size should be small relative to the population size. Specifically, the sample size should be less than 10% of the population size to ensure that the samples are approximately independent. This is often referred to as the "10% condition."

Underlying Population Distribution

The original population distribution does not need to be normal for the CLT to apply. The CLT works even if the population distribution is skewed, uniform, or has any shape.

However, the more skewed or irregular the population, the larger the sample size required to approximate a normal distribution of the sample mean.

23.  How does a probability distribution relate to the expected outcome of a random variable?
 - A probability distribution and the expected outcome of a random variable are closely related concepts in probability theory. The probability distribution describes how the probabilities are spread over the possible values of the random variable, and the expected outcome (or expected value) represents the weighted average of these values, taking into account their probabilities.
